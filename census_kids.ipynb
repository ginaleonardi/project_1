```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
clean = "clean_master_merge.csv"
clean_df = pd.read_csv(clean)
clean_df
```

```python
#cleaning n/a from feelings columns
#dropping duplicate columns
#renaming columns
clean_df = clean_df.dropna(subset=['ANXIOUS', 'WORRY', 'INTEREST', 'DOWN'])
clean_df = clean_df.drop(['SCRAM', 'RRACE.1','EEDUC.1'], axis=1)
clean_df = clean_df.rename(columns={'TENURE': 'Home Ownership',
                                    'EEDUC': 'Education',
                                    'RRACE': 'Race',
                                    'TBIRTH_YEAR': 'Birth Year',
                                    'RHISPANIC': 'Hispanic',
                                    'MS': 'Martial Status',
                                    'EGENID_BIRTH': 'Gender @ Birth',
                                    'GENID_DESCRIBE': 'Gender Id Describe',
                                    'SEXUAL_ORIENTATION': 'Sexual Orientation',
                                    'ANYWORK': 'Employment Status',
                                    'WEEK': 'Survey Date',
                                    'INCOME': 'Income',
                                    'KIDS_LT5Y': 'Kids 0-4',
                                    'KIDS_5_11Y': 'Kids 5-11',
                                    'KIDS_12_17Y': 'Kids 12-17',
                                    'EST_ST': 'State'})
clean_df
```

```python
#narrowing down to kids and feelings only
kids_df = clean_df.drop(['Education', 'Race','Birth Year', 'Hispanic', 'Martial Status', 'Gender @ Birth',
                         'Gender Id Describe', 'Sexual Orientation', 'Employment Status', 
                        'Survey Date', 'Income', 'Home Ownership', 'State'], axis=1)
kids_df
```

```python
#filling n/a with zeros and setting data type
kids_df["Kids 0-4"] = kids_df["Kids 0-4"].fillna(0)
kids_df["Kids 5-11"] = kids_df["Kids 5-11"].fillna(0)
kids_df["Kids 12-17"] = kids_df["Kids 12-17"].fillna(0)
kids_df = kids_df.astype(int)
kids_df
```

```python
#narrow down to one feeling
anxious = kids_df['ANXIOUS'].mean()
worry = kids_df['WORRY'].mean()
interest =kids_df['INTEREST'].mean()
down = kids_df['DOWN'].mean()
print(anxious, worry, interest, down)
```

```python
#sort out with no kids only
no_kids = kids_df.loc[(kids_df['Kids 0-4'] == 0) & (kids_df['Kids 5-11'] == 0) & (kids_df['Kids 12-17'] == 0), :]
no_kids
```

```python
anx_no_kids = no_kids["ANXIOUS"].value_counts()
anx_no_kids.plot(kind="bar")
plt.title("Anxiety Score with no Kids")
plt.xlabel("Anxiety Score")
plt.ylabel("Anxiety Score Count")
plt.xticks(rotation="vertical")
plt.savefig("Anxietynokids.png", bbox_inches='tight')
plt.show()
```

```python
#sorting to with kids only
with_kids = kids_df.loc[(kids_df['Kids 0-4'] == 1) | (kids_df['Kids 5-11'] == 1) | (kids_df['Kids 12-17'] == 1), :]
with_kids
```

```python
anx_with_kids = with_kids["ANXIOUS"].value_counts()
anx_with_kids.plot(kind="bar")
plt.title("Anxiety Score with Kids")
plt.xlabel("Anxiety Score")
plt.ylabel("Anxiety Score Count")
plt.xticks(rotation="vertical")
plt.savefig("Anxietywithkids.png", bbox_inches='tight')
plt.show()
```

```python
#narrow down further to highest reported feeling and greatest number
no_kids4 = no_kids.loc[no_kids["ANXIOUS"] == 4]
no_kids4
```

```python
#narrow down further to highest reported feeling and greatest number
with_kids4 = with_kids.loc[with_kids["ANXIOUS"] == 4]
with_kids4
```

```python
#make a new column that lumps all 3 kid questions columns into one
no_kids4['Kids'] = 0
```

```python
no_kids4
```

```python
#make a new column that lumps all 3 kid questions columns into one
with_kids4['Kids'] = 1
```

```python
with_kids4
```

```python
#merge both for plotting
final = pd.merge(no_kids4, with_kids4, how="outer", on=["Kids", "ANXIOUS"])
```

```python
final
```

```python
kidsy_n = final["Kids"].value_counts()
kidsy_n.plot(kind="pie", autopct="%1.1f%%", labels=["No Kids", "With Kids"])
plt.ylabel('Anxious 4 by Kids')
plt.savefig("Anxiety4kids.png", bbox_inches='tight')
plt.show()
```
